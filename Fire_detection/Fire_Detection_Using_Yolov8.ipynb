```python
from ultralytics import YOLO
import cvzone
import cv2
import math

# Running real time from webcam
cap = cv2.VideoCapture('Fall.jpg')
model = YOLO('fire1.pt')

# Reading the classes
classnames = ['fire']

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (640, 480))
    result = model(frame, stream=True)

    # Getting bbox, confidence, and class names information to work with
    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            Class = int(box.cls[0])
            if confidence > 50:
                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 5)
                cvzone.putTextRect(frame, f'{classnames[Class]} {confidence}%', [x1 + 8, y1 + 100],
                                   scale=1.5, thickness=2)

    cv2.imshow('frame', frame)

    # Wait for a key event for 1 millisecond
    key = cv2.waitKey(1) & 0xFF

    # Check if the 'q' key is pressed
    if key == ord('q'):
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```

```python
import streamlit as st
import cv2
from PIL import Image
import numpy as np
from ultralytics import YOLO
import cvzone
import math

# Load YOLO model
model = YOLO('fire1.pt')

# Function to perform fire detection
def detect_fire(image):
    image_np = np.array(image)
    result = model(image_np, size=640)
    
    # Getting bbox, confidence, and class names information to work with
    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            Class = int(box.cls[0])
            if confidence > 50:
                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                cv2.rectangle(image_np, (x1, y1), (x2, y2), (0, 0, 255), 5)
                cvzone.putTextRect(image_np, f'Fire {confidence}%', [x1 + 8, y1 + 30],
                                   scale=1.0, thickness=2)

    return image_np

# Streamlit app
def streamlit_app():
    st.title("Fire Detection App")

    uploaded_file = st.file_uploader("Choose a file", type=["jpg", "jpeg", "png"])

    if uploaded_file is not None:
        image = Image.open(uploaded_file)

        st.image(image, caption="Uploaded Image", use_column_width=True)

        if st.button("Detect Fire"):
            with st.spinner('Detecting fire...'):
                result_image = detect_fire(image)
                st.image(result_image, caption="Result", use_column_width=True)

# Run the Streamlit app in the notebook
!streamlit run streamlit_app.py
```
